```python
#Function to clean up NC values
def enroll_Replace (df):
    #replace NC values with zero
    df.replace('NC', '0', inplace=True)
    #replace <10 value with zero
    df.replace('^.*<.*$', '0', regex=True, inplace=True)
    #create a new column for enrollment total and use iloc to add
    df[f'Enrollment_Total'] = df.iloc[:, -17:-1].sum(1)
     #create list of columns to drop, then drop
    cols = df.columns
    #force them to be integers
    df[cols] = df[cols].apply(pd.to_numeric, errors='coerce')
   
   
    #drop unnecessary columns
    return pd.DataFrame({"Building_Name": df["Building Name"],"Building_IRN": df["Building IRN"], "District_IRN": df["District IRN"], f'Enrollment_Total': df[f'Enrollment_Total']})
```
